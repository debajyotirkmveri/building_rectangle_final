```python
import json
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# Path to the JSON file
json_file_path = "/home/sysadm/Documents/Project/Annotation_rectangle/Post_Event_115.json"

# Load JSON data from the file
with open(json_file_path, 'r') as file:
    data = json.load(file)

# Load the image
image_path = "/home/sysadm/Documents/Project/Post_Event_Grids_In_JPEG/Post_Event_115.jpg"

# Load the image
image = Image.open(image_path)

# Create figure and axes
fig, ax = plt.subplots()

# Display the image
ax.imshow(image)

# Define colors for each label
label_colors = {
    "undamagedcommercialbuilding": "blue",
    "damagedcommercialbuilding": "red",
    "undamagedresidentialbuilding": "green",
    "damagedresidentialbuilding": "yellow"
}

# Plot each bounding box
for shape in data["shapes"]:
    label = shape["label"]
    points = shape["points"]
    color = label_colors.get(label, "black")  # Default to black if label not found in dictionary
    if shape["shape_type"] == "rectangle":
        # Extract rectangle coordinates
        x_min, y_min = points[0]
        x_max, y_max = points[1]
        width = x_max - x_min
        height = y_max - y_min
        # Plot rectangle
        ax.add_patch(patches.Rectangle((x_min, y_min), width, height, edgecolor=color, fill=False))

# Save the annotated image
annotated_image_path = "/home/sysadm/Documents/Project/post_event_115_annotated.jpg"
plt.savefig(annotated_image_path)

# Show plot (optional)
plt.show()
```

```python
import json
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# Input folder containing JSON files
input_folder = "/home/sysadm/Documents/Project/sample_image/json_file"

# Output folder to save images
output_folder = "/home/sysadm/Documents/Project/sample_image/annoattion"

# Define colors for each label
label_colors = {
    "undamagedcommercialbuilding": "blue",
    "damagedcommercialbuilding": "red",
    "undamagedresidentialbuilding": "green",
    "damagedresidentialbuilding": "yellow"
}

# Iterate over each file in the folder
for filename in os.listdir(input_folder):
    if filename.endswith('.json'):
        # Load JSON data from the file
        json_file_path = os.path.join(input_folder, filename)
        with open(json_file_path, 'r') as file:
            data = json.load(file)

        # Load the image
        image_path = data["imagePath"]
        if image_path.startswith("../"):
            image_path = image_path[1:]
        image = Image.open(image_path)

        # Create figure and axes
        fig, ax = plt.subplots()

        # Display the image
        ax.imshow(image)

        # Plot each bounding box
        for shape in data["shapes"]:
            label = shape["label"]
            points = shape["points"]
            color = label_colors.get(label, "black")  # Default to black if label not found in dictionary
            if shape["shape_type"] == "polygon":
                # Convert points to matplotlib-compatible format
                x = [point[0] for point in points]
                y = [point[1] for point in points]
                # Plot polygon
                ax.add_patch(patches.Polygon(xy=list(zip(x, y)), edgecolor=color, fill=False))
            elif shape["shape_type"] == "rectangle":
                # Extract rectangle coordinates
                x, y = points[0]
                width = points[1][0] - x
                height = points[1][1] - y
                # Plot rectangle
                ax.add_patch(patches.Rectangle((x, y), width, height, edgecolor=color, fill=False))

        # Save the image with bounding boxes to the output folder
        output_file_path = os.path.join(output_folder, filename[:-5] + ".jpg")  # Remove '.json' extension
        plt.savefig(output_file_path)
        plt.close()  # Close the plot to release memory

print("Images with bounding boxes saved to", output_folder)
```

```python
import json
import os
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# Input folder containing JSON files
input_folder = "/home/sysadm/Documents/Project/sample_image/json_file"

# Output folder to save images
output_folder = "/home/sysadm/Documents/Project/sample_image/annotation1"

# Define colors for each label
label_colors = {
    "undamagedcommercialbuilding": "blue",
    "damagedcommercialbuilding": "red",
    "undamagedresidentialbuilding": "green",
    "damagedresidentialbuilding": "yellow"
}

# Define abbreviated class names
abbreviated_classes = {
    "undamagedcommercialbuilding": "udcb",
    "damagedcommercialbuilding": "dcb",
    "undamagedresidentialbuilding": "udrb",
    "damagedresidentialbuilding": "drb"
}

# Iterate over each file in the folder
for filename in os.listdir(input_folder):
    if filename.endswith('.json'):
        # Load JSON data from the file
        json_file_path = os.path.join(input_folder, filename)
        with open(json_file_path, 'r') as file:
            data = json.load(file)

        # Load the image
        image_path = data["imagePath"]
        if image_path.startswith("../"):
            image_path = image_path[1:]
        image = Image.open(image_path)

        # Create figure and axes
        fig, ax = plt.subplots()

        # Display the image
        ax.imshow(image)

        # Plot each bounding box
        for shape in data["shapes"]:
            label = shape["label"]
            points = shape["points"]
            color = label_colors.get(label, "black")  # Default to black if label not found in dictionary
            if shape["shape_type"] == "polygon":
                # Convert points to matplotlib-compatible format
                x = [point[0] for point in points]
                y = [point[1] for point in points]
                # Plot polygon
                ax.add_patch(patches.Polygon(xy=list(zip(x, y)), edgecolor=color, fill=False))
            elif shape["shape_type"] == "rectangle":
                # Extract rectangle coordinates
                x, y = points[0]
                width = points[1][0] - x
                height = points[1][1] - y
                # Plot rectangle
                ax.add_patch(patches.Rectangle((x, y), width, height, edgecolor=color, fill=False))
                
                # Add abbreviated class name as text
                cls_text = abbreviated_classes.get(label, label)
                ax.text(x, y, cls_text, fontsize=8, color='white', verticalalignment='top', bbox=dict(facecolor=color, alpha=0.5))

        # Save the image with bounding boxes and text annotations to the output folder
        output_file_path = os.path.join(output_folder, filename[:-5] + ".jpg")  # Remove '.json' extension
        plt.savefig(output_file_path)
        plt.close()  # Close the plot to release memory

print("Images with bounding boxes and text annotations saved to", output_folder)
```
